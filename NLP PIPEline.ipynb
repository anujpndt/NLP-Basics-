```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/NLP PIPELINE/IMDB Dataset.csv")
df.head()
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

def preprocess_text(text):
    doc = nlp(text.lower())
    lemmatized_words = [token.lemma_ for token in doc if not token.is_stop and token.is_alpha]
    return " ".join(lemmatized_words)

# Example
text = "Hii!!! Welcome to NLP 😊. Check this out: https://example.com"
cleaned = preprocess_text(text)
print(cleaned)
```

```python
df['clean_review']=df['review'].apply(preprocess_text)
df.head()
```

```python
# Problem 2

# Find out the number of words in the entire corpus and also the total number of unique words(vocabulary) using just python
# Combine all reviews into one large text corpus
full_corpus = " ".join(df["review"].astype(str))

# Tokenize the corpus into words (basic splitting)
words = full_corpus.split()

# Find total number of words
total_words = len(words)

# Find the number of unique words (vocabulary size)
unique_words = len(set(words))

print("Total number of words in the corpus:", total_words)
print("Total number of unique words (vocabulary size):", unique_words)

```

```python
# Convert Sentiment Labels ('positive', 'negative') into numeric form

import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import CountVectorizer

label_encoder = LabelEncoder()
df["sentiment_encoded"] = label_encoder.fit_transform(df["sentiment"])  # 'positive' → 1, 'negative' → 0

# Apply One-Hot Encoding to Text Data
vectorizer = CountVectorizer(binary=True)  # Convert words into binary one-hot encoding
X = vectorizer.fit_transform(df["review"])

# Convert to DataFrame
one_hot_encoded_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# Display first few rows
print(one_hot_encoded_df.head())
```
